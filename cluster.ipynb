```python
from sklearn.datasets import load_iris
import pandas as pd
iris = load_iris()
df = pd.DataFrame(iris["data"], columns=iris["feature_names"])
df["ans"] = iris["target"]
# df.to_csv("iris.csv", encoding="utf-8", index=False)
df
```

```python
from sklearn.cluster import KMeans
clu = KMeans(n_clusters=3)
clu.fit(iris["data"])
```

```python
# 分群以後對每個資料打上的標籤
df["cluster_labels"] = clu.labels_
df
```

```python
# 分群以後的三個中心點
clu.cluster_centers_
```

```python
# 為了畫圖, 我重新分群一次(只用兩個特徵)
new = df[ ["sepal length (cm)", "petal length (cm)"] ]
clu = KMeans(n_clusters=3)
clu.fit(new)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# plt.show()
# 偷懶指令: 只要那格有圖, 就自動show
%matplotlib inline
x = df["sepal length (cm)"]
y = df["petal length (cm)"]
h = clu.labels_
sns.scatterplot(x, y, hue=h)
```

```python
from sklearn.metrics import silhouette_score
xs = []
ys = []
for k in range(2, 11):
    clu = KMeans(n_clusters=k)
    clu.fit(iris["data"])
    xs.append(k)
    ys.append(silhouette_score(iris["data"], clu.labels_))
print("K:", xs)
print("親疏程度(Silh)")
```
